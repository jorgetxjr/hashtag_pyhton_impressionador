```python
import pyautogui as auto
import pyperclip
import time

```

```python
class MeuRobo():
    def __init__(self, tempo_espera):
        auto.PAUSE = 1
        self.tempo_espera = tempo_espera

    def abrir_navegador(self,navegador:str):
        auto.press("win")
        auto.write(navegador)
        auto.press("enter")

    def abrir_site(self,site:str):
        pyperclip.copy(site)
        #auto.hotkey("ctrl","l")
        auto.hotkey("ctrl","v")
        auto.press("enter")
        print("Aguardando o site carregar")
        time.sleep(self.tempo_espera)
        print("OK")

    def clicar(self,x,y,botao):
        auto.click(x,y,button=botao)
    
    def identificar_posicao(self):
        for i in range(5):
            print(f'Pegando posição e {5-i} segundos')
            time.sleep(1)

        print(auto.position().x,auto.position().y)
```

```python
t800 = MeuRobo(3)
# t800.abrir_navegador("brave")
# t800.abrir_site("https://www.hashtagtreinamentos.com/blog")
t800.identificar_posicao()
```
