### Puxar base de dados

```python
import pandas as pd

tabela = pd.read_csv('exportacoes_franca.csv')
display(tabela)
```

### Informações Gerais

Quantas linhas têm a tabela? Existe algum valor vazio que deveria ser tratado?

```python
display(tabela.info())
```

```python
#Função de formatação

def formatar(valor):
    valor_formatado = f'US${valor:,.2f}'
    return valor_formatado
```

### Como foi a evolução das exportações para a frança ao longo dos anos?

Crie uma tabela com os anos de 2016, 2017, 2018, 2019 e 2020 e o valor US$ de exportações

```python
tabela_anos = tabela[["Year", "US$ FOB"]].groupby("Year").sum()
tabela_anos = tabela_anos['US$ FOB'].apply(formatar)
display(tabela_anos)
```

### Quais os produtos mais exportados ao longo de todo o período?

Queremos saber os produtos com maior valor US$ de exportação

```python
produtos_exportados = tabela[["SH2 Description","US$ FOB"]].groupby("SH2 Description").sum()
produtos_exportados = produtos_exportados.sort_values(by='US$ FOB', ascending=False)
display(produtos_exportados)
```

### Em 2020 qual cidade mais exportou para a França?
Em valores US$

```python
condicao = tabela['Year']==2020
export_2020 = tabela.loc[condicao,:]
tabela_cidades = export_2020[['City', 'US$ FOB']].groupby("City").sum()
tabela_cidades = tabela_cidades.sort_values(by='US$ FOB', ascending=False)
display(tabela_cidades)
```

### Quais os produtos mais exportados (em US$) que as 2 maiores cidades (em exportação em 2020) exportaram?

```python
for i,cidade in enumerate(tabela_cidades.index):
    if i == 2:
        break

    tabela_cidade = export_2020.loc[export_2020['City']==cidade,:]
    tabela_produtos_cidade = tabela_cidade[['SH2 Description','US$ FOB']].groupby('SH2 Description').sum()
    tabela_produtos_cidade = tabela_produtos_cidade.sort_values('US$ FOB', ascending=False)
    print(f"Exportações da cidade {cidade}:")
    display(tabela_produtos_cidade)
```
